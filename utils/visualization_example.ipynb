```python
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from visualization import plot_decision_region
```

## Data Preparation

```python
# use PCA to extract 2 components from X
X, y = load_iris(return_X_y=True)
pca = PCA(n_components=2)
X_transformed = pca.fit_transform(X)
pca.explained_variance_ratio_
```

## Plot Decision Region

```python
# SVM
svc = SVC()
plot_decision_region(X_transformed, y, svc, xlabel='pca0', ylabel='pca1')
```

```python
# logistic regression
lr = LogisticRegression(C=2.0)
plot_decision_region(X_transformed, y, lr, xlabel='pca0', ylabel='pca1')
```
