```python
from feature_selection import SBS
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.decomposition import PCA, KernelPCA
from visualization import plot_decision_region
from sklearn.svm import SVC
from sklearn.datasets import load_iris
```

```python
# use SBS, LDA, PCA, KernelPCA to extract 2 components from X,
# and deploy SVC to make decision region
X, y = load_iris(return_X_y=True)
svc = SVC()

sbs = SBS(svc, n_components=2)
lda = LDA(n_components=2)
pca = PCA(n_components=2)
kpca_rbf = KernelPCA(n_components=2, kernel='rbf')
kpca_linear = KernelPCA(n_components=2, kernel='linear')

X_transformed_sbs = sbs.fit_transform(X, y) # supervised
X_transformed_lda = lda.fit_transform(X, y) # supervised
X_transformed_pca = pca.fit_transform(X) # unsupervised
X_transformed_kpca_rbf = kpca_rbf.fit_transform(X) # unsupervised
X_transformed_kpca_linear = kpca_linear.fit_transform(X) # unsupervised
```

## Plot Decision Region

```python
# SBS
plot_decision_region(X_transformed_sbs, y, svc, xlabel='pca0', ylabel='pca1')
```

```python
# LDA
plot_decision_region(X_transformed_lda, y, svc, xlabel='pca0', ylabel='pca1')
```

```python
# PCA
plot_decision_region(X_transformed_pca, y, svc, xlabel='pca0', ylabel='pca1')
```

```python
# KernelPCA(rbf kernel)
plot_decision_region(X_transformed_kpca_rbf, y, svc, xlabel='pca0', ylabel='pca1')
```

```python
# KernelPCA(linear kernel)
plot_decision_region(X_transformed_kpca_linear, y, svc, xlabel='pca0', ylabel='pca1')
```
